```python
dataframe = pd.read_csv('/content/gdrive/MyDrive/Bigdata/data/penguins_size.csv')
dataframe.head()
```

```python
dataframe.shape
```

```python
dataframe = dataframe.dropna()
dataframe.shape
```

```python
dataframe = dataframe.drop('sex', axis=1)
dataframe.head()
```

```python
data = dataframe.values
```

```python
data
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
X = data[:,1:]
y = data[:,0]
```

```python
X
```

```python
y
```

```python
from collections import Counter
counter = Counter()
counter.update(y)
print(counter)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for i in range(X.shape[1]):
    X[:,i] = encoder.fit_transform(X[:,i])
y=encoder.fit_transform(y)
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
X = minmax.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print('Train:')
print(X_train.shape)
print(y_train.shape)
print('Test:')
print(X_test.shape)
print(y_test.shape)
```

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train)
print(y_train[:2])
```

```python
from keras.models import Sequential
model1 = Sequential()
```

Model

```python
from tensorflow.keras import layers
# model 1
model1.add(layers.Dense(16, activation="tanh"))
model1.add(layers.Dense(32, activation="tanh"))      
model1.add(layers.Dense(3, activation="softmax")) 
model1.compile(loss="categorical_crossentropy", optimizer="adam",metrics=["accuracy"])
```

```python
model1.fit(X_train,y_train, epochs=50, batch_size=10,verbose=2)
```

```python
y_pred = model1.predict(X_test)
```

```python
y_pred = y_pred.argmax(axis=1)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
# model 2
model2 = Sequential()
model2.add(layers.Dense(64, input_dim=5, kernel_initializer="lecun_normal", activation="relu"))
model2.add(layers.Dense(32, activation="relu"))   
model2.add(layers.Dense(16, activation="sigmoid"))      
model2.add(layers.Dense(3, activation="softmax")) 
model2.compile(loss="categorical_crossentropy", optimizer="adam",metrics=["accuracy"])
model2.fit(X_train,y_train, epochs=50, batch_size=10,verbose=2)
y_pred_2 = model2.predict(X_test)
y_pred_2 = y_pred_2.argmax(axis=1)
print(classification_report(y_test,y_pred_2))
```
